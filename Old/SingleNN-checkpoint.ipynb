```python
import MagicTools as mt

import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout

import numpy as np

import matplotlib.pyplot as plt

from scipy.signal import savgol_filter

import random
random.seed(1)
```

```python
rdflist = []
potlist = []
for i in range(1, 31):
    for j in range(10):
        rdf = mt.ReadRDF('RDFs/{}bead{}.rdf'.format(i,j), quiet = True)
        pot = mt.ReadPot('MagiC_Potentials/{}beadpotential{}.pot'.format(i,j), quiet = True)
        rdflist.append(rdf)
        potlist.append(pot)
```

```python
x = np.linspace(0.025, 14.975, 300)
for i in range(len(x)):
    x[i] = round(x[i], 3)
```

```python
rdfs = []
for i in range(len(rdflist)):
    N = len(rdflist[i].DFs)
    
    for j in range(N):
        holder = np.empty((300))
        startindex = np.where(x == round(rdflist[i].DFs[j].x[0], 3))[0][0]
        for k in range(300):
            if k < startindex:
                holder[k] = 0
            elif k >= startindex and k-startindex < len(rdflist[i].DFs[j].x):
                holder[k] = rdflist[i].DFs[j].y[k-startindex]
            else:
                holder[k] = 0
                
        holder = tf.constant(holder)
        rdfs.append(holder)
```

```python
pots = []
for i in range(len(rdflist)):
    N = len(rdflist[i].DFs)
    
    for j in range(N):
        holder = np.empty((300))
        
        startindex = np.where(x == round(potlist[i].DFs[j].x[0], 3))[0][0]
        for k in range(300):
            if k < startindex:
                holder[k] = 300 
            elif k >= startindex and k-startindex < len(potlist[i].DFs[j].x):
                holder[k] = potlist[i].DFs[j].y[k-startindex] + 30
            else:
                print("error", i, j)
        holder = tf.constant(holder)
        pots.append(holder)
```

```python
inputs = tf.stack(rdfs)
outputs = tf.stack(pots)
```

```python
def mpe(ytrue, ypred):
    loss = tf.keras.losses.mean_absolute_percentage_error(ytrue,ypred)
    return loss

def msle(ytrue, ypred):
    loss = tf.keras.losses.mean_squared_logarithmic_error(ytrue,ypred)
    return loss
```
